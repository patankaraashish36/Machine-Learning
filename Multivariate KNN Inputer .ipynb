```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer , SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Most accuracy_score Technique =  constant , fill_value = 99  (SimpleImputer)

```python
df = pd.read_csv("C:/Users/patan/DS/Machine Learning/Dataset/train.csv", usecols = ['Age','Fare','Survived'])
print(df.shape)
df.head()
```

```python
X = df.drop(columns = "Survived") 
```

```python
y = df["Survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.2,shuffle=False )  # shuffle=False is Used for the split data into to 
X_train     # two Sequatial part and random_state split into two random part 
```

```python
y_test
```

#  KNN Imputation

```python
knn  = KNNImputer(n_neighbors=6 )   # add_indicator=True
```

```python
X_train_ft = knn.fit_transform(X_train)
X_test_ft = knn.transform(X_test)
X_train_ft
```

```python
new_df = pd.DataFrame(X_train_ft , columns = ["Imputed_Age" , "Imputed_Fare" ])
# new_df = pd.DataFrame(X_train_ft , columns = ["Imputed_Age" , "Imputed_Fare" ,"Missing" ])
```

```python
new_df
```

```python
lr = LogisticRegression()
lr.fit(X_train_ft, y_train)
y_prediction_1 =lr.predict(X_test_ft)
accuracy_score(y_test,y_prediction_1)
```

```python
# accuracy_score =  0.6558988764044944
```

#  Simpule imputer accuracy_score  (Mean)

```python
si = SimpleImputer()  # add_indicator=True
```

```python
X_train_si = si.fit_transform(X_train)
X_test_si = si.transform(X_test)
```

```python
X_train_si
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train_si, y_train)
y_prediction_2 = lr.predict(X_test_si)
accuracy_score(y_test, y_prediction_2)
```

```python
# accuracy_score = 0.7039106145251397
```

#  Simpule imputer accuracy_score (Median)

```python
si = SimpleImputer(strategy = "median")  # add_indicator=True
```

```python
X_train_si2 = si.fit_transform(X_train)
X_test_si2 = si.transform(X_test)
X_train_si2
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train_si2,y_train)
```

```python
y_pred_3  = lr.predict(X_test_si2)
```

```python
accuracy_score(y_test, y_pred_3)
```

```python
# accuracy_score = 0.7039106145251397
```

#  Simpule imputer accuracy_score (most_frequent)

```python
si =  SimpleImputer(strategy = "most_frequent") # add_indicator=True
si
```

```python
X_train_si3 = si.fit_transform(X_train)
X_test_si3 = si.transform(X_test)
```

```python
lr = LogisticRegression()
lr.fit(X_train_si3,y_train)
y_pred_4 = lr.predict(X_test_si3)
accuracy_score(y_test ,y_pred_4)
```

```python
# accuracy_score = 0.7039106145251397
```

#  Simpule imputer accuracy_score (constant)

```python
si = SimpleImputer(strategy = "constant" , fill_value = -1 ) # add_indicator=True
```

```python
X_train_si4  = si.fit_transform(X_train)
X_test_si4 = si.transform(X_test)
```

```python
lr = LogisticRegression()
lr.fit(X_train_si4 , y_train)
y_pred_5 = lr.predict(X_test_si4)
accuracy_score(y_test,y_pred_5 )
```

```python
# accuracy_score = 0.7039106145251397    fill_value = 99 == 0.7150837988826816
```
